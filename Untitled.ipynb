## Párová analýza dát

### Spojitý - spojitý

#### Zistenie vzťahov, korelácií medzi atribútmi

Aj napriek tomu, že atribút class je reprezentovaný číslom, vynechám ho z korelačnej tabuľky, kedže ide o kategorický atribút

```python
train_data.drop(['class'], axis=1).corr()
```

Z korelačnej tabuľky môžeme vyčítať, že jediná dvojica atribútov, ktorá má silnú koreláciu je ,,skewness_glucose" a ,,kurtosis_glucose".

```python
plt.scatter(train_data.skewness_glucose, train_data.kurtosis_glucose)
```

Naozaj na grafe môžeme vidieť pozitívnu koreláciu medzi atribútmi skewness_glucose a kurtosis_glucose

### Spojitý - kategorický

Pomocou tohto grafu chceme zistiť, či cieľový atribút ,,class" tvorí zhluky v závislosti od dvoch iných atribútov.

```python
sns.pairplot(train_data, hue="class")
```

Napriek tomu že zhluky nie sú nikde dokonalé, v podgrafoch s atribútmi ,,skewness_glucose" a ,,kurtosis_glucose" môžeme vidieť, že pacienti u ktorých bola cukrovka zistená mali vyššie hodnoty týchto atribútov.

#### Vizualizácia pomocou boxplotu a violinplotu

```python
sns.boxplot(x='class', y='skewness_glucose', data=train_data)
```

```python
sns.boxplot(x='class', y='kurtosis_glucose', data=train_data)
```

Na grafoch môžeme naozaj vidieť, to čo sme vyčítali z pairplot-u vyššie. Pri atribúte kurtosis_glucose je rozdiel hodnôt u ,,cukrovkárov" a zdravých ľudí výraznejší ako pri atribúte skewness_glucose

```python
sns.violinplot(x='class', y='skewness_glucose', data=train_data)
```

```python
sns.violinplot(x='class', y='kurtosis_glucose', data=train_data)
```

Na týchto violinplot-och môžeme vidieť že u zdravých pacientov sa hodnoty kurtosis_glucose a skewness_glucose pohybujú okolo 0, zatiaľ čo u pacientov s cukrovkou sú tieto hodnoty takmer rovnomerne rozptýlené na dlhšom intervale.

```python
sns.boxplot(x='class', y='age', data=train_data)
```

```python
sns.violinplot(x='class', y='age', data=train_data)
```

### Kategorický - kategorický

V tejto časti sme sa snažili zobraziť vzťahy a nájsť prípadné závislosti medzi cieľovým atribútom class a ďalšími kategorickými atribútmi.

```python
pd.crosstab(index=train_data["class"], columns=train_data["race"])
```

```python
pd.crosstab(index=train_data["class"], columns=train_data["sex"])
```

```python
pd.crosstab(index=train_data["class"], columns=train_data["marital-status"])
```

```python
pd.crosstab(index=train_data["class"], columns=[train_data["race"], train_data["sex"]])
```

Nepodarilo sa nám nájsť žiadnu signifikantnú závislosť. V každej skúmanej kategórií sú ,,cukrovkári" a zdraví ľudia približne v pomere 1:3

## Identifikácia problémov v dátach spolu s predpokladaným scenárom riešenia v ďalšej fáze: 
- mean_glucose je typu object, zatiaľ čo std_glucose, skewness_glucose, kurtosis_glucose sú typu float. Majú teda nejednotné typy a ťažko sa analyzujú
- v datasete personal_train sa nachádza 3933 jedinečných mien (pacientov), v druhom datasete je ich o 50 viac. Identifikovali sme 50 duplikátov mien. 
- V datasete other_train sa ako názvy atribútov nachádzajú pomlčky (-) namiesto podtržníkov (_). Je to nielen nejednotný formát, ale názvy týchto stĺpcov nie je možné analyzovať potrebnými funkciami. V datasete sa nachádza viac atribútov s týmto problémom.
- V stĺpci pre vek (age) sa nachádzajú hodnoty ako -1 alebo 113. Sú to veľmi vychýlené hodnoty a navyše vek -1 nie je možný.
- Stĺpec class obsahuje aj NAN hodnoty - nie je možné vedieť skutočný stav pacienta
- V stĺpci "medical info" sa nachádza veľa hodnôt, ktoré by bolo možné rozdeliť do osobitných stĺpcov
