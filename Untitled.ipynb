## Spojenie (merge) datasetov personal_train a other_train

Dva datasety, ktoré sme mali k dispozícii - "personal_train" a "other_train" sme spojili na základe mena a adresy do jedného datasetu.

## Identifikácia problémov v dátach spolu s predpokladaným scenárom riešenia v ďalšej fáze: 
- mean_glucose je typu object, zatiaľ čo std_glucose, skewness_glucose, kurtosis_glucose sú typu float. Majú teda nejednotné typy a ťažko sa analyzujú
- v datasete personal_train sa nachádza 3933 jedinečných mien (pacientov), v druhom datasete je ich o 50 viac. Identifikovali sme 50 duplikátov mien. 
- V datasete other_train sa ako názvy atribútov nachádzajú pomlčky (-) namiesto podtržníkov (_). Je to nielen nejednotný formát, ale názvy týchto stĺpcov nie je možné analyzovať potrebnými funkciami. V datasete sa nachádza viac atribútov s týmto problémom.
- V stĺpci pre vek (age) sa nachádzajú hodnoty ako -1 alebo 113. Sú to veľmi vychýlené hodnoty a navyše vek -1 nie je možný.
- Stĺpec class obsahuje aj NAN hodnoty - nie je možné vedieť skutočný stav pacienta
- V stĺpci "medical info" sa nachádza veľa hodnôt, ktoré by bolo možné rozdeliť do osobitných stĺpcov
